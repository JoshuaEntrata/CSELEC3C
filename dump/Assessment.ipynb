```python
fao_null_values = {}

for column in fao_df.columns:
    if fao_df[column].isnull().sum() > 0:
        fao_null_values[column] = fao_df[column].isnull().sum()
    
fao_null_values
```

```python
num_unique_vals = {}

for column in fao_df.columns:
    num_unique_vals[column] = fao_df[column].nunique()
    
num_unique_vals
```
