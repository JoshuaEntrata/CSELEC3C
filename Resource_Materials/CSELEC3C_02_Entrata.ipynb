```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('pokemon.csv', index_col=0, encoding='ISO-8859-1')
```

```python
df1.head()
```

```python
df1.describe()
```

```python
g = plt.hist(df1['Speed'], histtype='bar', ec='black')
g = plt.xlabel('Pokemon Speed')
g = plt.ylabel('Frequency')
plt.show()
```

```python
bins_edges = [x for x in range(0, 151, 10)]
g = plt.hist(df1['Speed'], histtype='bar', ec='black', bins=bins_edges)
g = plt.xlabel('Pokemon Speed')
g = plt.ylabel('Frequency')
plt.show()
```

```python
sns.set_style()
sns.histplot(df1.Speed)
```

```python
sns.lmplot(x='Attack', y='Defense', data=df1)
```

```python
sns.lmplot(x='Attack', y='Defense', data=df1, fit_reg=False)
```

```python
sns.lmplot(x='Attack', y='Defense', data=df1, fit_reg=False, hue='Generation')
```

```python
g = sns.catplot(
    x='Type 1',
    y='Attack',
    data=df1,
    hue='Generation',
    col='Generation'
)
g.set_xticklabels(rotation=-45)
```

```python
stats_df = df1.drop(['Total', 'Generation', 'Legendary'], axis=1)
sns.set_style('whitegrid')
sns.boxplot(data=stats_df)
```

```python
f = sns.violinplot(x='Type 1', y='Attack', data=df1)
f.set_xticklabels(rotation=-45)
```

```python
sns.swarmplot(
    x='Type 1',
    y='Attack',
    data=df1
)
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(
    x='Type 1',
    y='Attack',
    data=df1,
    inner=None
)
sns.swarmplot(
    x='Type 1',
    y='Attack',
    data=df1,
    color='k',
    alpha=0.7
)
plt.title('Attack by Type')
```
